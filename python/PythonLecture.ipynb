## Containers

```python
colors = ['red', 'blue', 'green', 'black', 'white']
type(colors)     
```

```python
colors[2]
```

```python
colors[-1]
```

**Slicing syntax: colors[start:stop:stride]**

```python
colors[0] = 'yellow'
colors

colors[2:4] = ['gray', 'purple']
colors
```

```python
# Reverse

rcolors = colors[::-1]
rcolors

rcolors2 = list(colors)
rcolors2

rcolors2.reverse() # in-place
rcolors2

# Concatenate
rcolors + colors

rcolors * 2

# Sort
sorted(rcolors) # new object

rcolors

rcolors.sort()  # in-place
rcolors
```
